##                                             Topics: Inheritance and Polymorphism

```python
# ##  1.                      Implement the  parent- "Sportsman" and child- "Cricketer", "Footballer"

#  classes. The details of attributes and methods are as below:

#  # ---------------------------------------------------------------------------------------------------------------------

#  ##                                                   "Sportsman" : Attributes

#  sport - stores the name of sport
#  role  - stores the role/position of player
#  wins  - stores the no. of games/matches won by player

#  ##                                                    "Sportsman" : Methods

#  Constructor to initialize : sport, role and wins.

#  plays() - prints the name of the sport played, ie value inside ‘sport’
#  stats() - prints the number of wins given by wins attribute, as ‘Player has ___ wins’

#   # ---------------------------------------------------------------------------------------------------------------------

#  ##                                                   "Cricketer" : Attributes

#  sport, role, wins are automatically inherited from "Sportsman".

#  runs     - stores number of runs scored by player
#  catches  - stores number of catches by player

#  ##                                                    "Cricketer" : Methods

#  Constructor to initialize ->  runs, catches and call to parent class constructor to initialize sport, role wins.

#  plays()- No change in method definitions. Same as Sportsman.
#  stats()- No change in method definition. Same as Sportsman

#   # ---------------------------------------------------------------------------------------------------------------------

#  ##                                                    "Footballer" : Attributes

#  sport, role, wins are automatically inherited from "Sportsman".

#  goals      - stores number of goals scored/blocked by player
#  use_hands  - Stores ‘yes’ for ‘goalkeeper’ role and ‘no’ for all others

#  ##                                                     "Footballer" : Methods

#  Constructor to initialize goals, use_hands and call to parent class constructor to initialize sport, role, wins.

#  plays() - No change in method definitions. Same as Sportsman
#  stats() - Change in method definition. It now prints the value in goals as follows

#  # ---------------------------------------------------------------------------------------------------------------------


#                                      ‘The number of goals scored is : __’ if role is not Goalkeeper.
#                                     Otherwise prints ‘The number of goals blocked is: __’
#                                     display_role()- prints the value in role

#   # ---------------------------------------------------------------------------------------------------------------------
```

```python
# Solution :

# ------------------------------------------------------------------------------------------------------------------------------------

class Sportsman:

    def __init__(self, sport, role, wins):

        self.sport = sport
        self.role = role
        self.wins = wins

    def plays(self):

        print(f"The sport played is: {self.sport}")

    def stats(self):

        print(f"Player has {self.wins} wins")

# ---------------------------------------------------------------------------------------------------------------------------------------

class Cricketer(Sportsman):

    def __init__(self, sport, role, wins, runs, catches):

        super().__init__(sport, role, wins)

        self.runs = runs
        self.catches = catches

# ---------------------------------------------------------------------------------------------------------------------------------------


class Footballer(Sportsman):

    def __init__(self, sport, role, wins, use_hands, goals):

        super().__init__(sport, role, wins)

        self.goals = goals
        self.use_hands = use_hands

    def stats(self):

        if self.role.lower() == 'goalkeeper':

            print(f"The number of goals blocked is: {self.goals}")

        else:

            print(f"The number of goals scored is: {self.goals}")

    def display_role(self):

        print(f"The role is: {self.role}")

# ---------------------------------------------------------------------------------------------------------------------------------------


#                                                   Execute the code and verify the output

Player1 = Cricketer('Cricket', 'Batsman', 145, 0, 1)
Player2 = Footballer('Football', 'Goalkeeper', 5, 'yes', 2)
Player3 = Footballer("Football", 'Forward Player', 3, 'no', 1)

# ---------------------------------------------------------------------------------------------------------------------------------------


Player1.plays()
Player2.plays()

print("\n -------------------------------------------------------------------------------------------------------------------")

Player1.stats()
Player2.stats()
Player3.stats()

print("\n -------------------------------------------------------------------------------------------------------------------")

# ---------------------------------------------------------------------------------------------------------------------------------------

Player2.display_role()

print("\n -------------------------------------------------------------------------------------------------------------------")


# ---------------------------------------------------------------------------------------------------------------------------------------

```
